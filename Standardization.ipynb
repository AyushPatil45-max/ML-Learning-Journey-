# Standardization

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('Social_Network_Ads.csv')
```

```python
df.sample(5)
```

```python
df=df.iloc[:,2:]
```

```python
df
```

### Train and Test split of the data

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(df.drop('Purchased',axis=1),df['Purchased'],test_size=0.3,random_state=0)
x_train.shape,x_test.shape
```

```python
x_train.head(3)
```

```python
from sklearn.preprocessing import StandardScaler

scaler=StandardScaler()
scaler.fit(x_train)

x_train_scaled=scaler.transform(x_train)
x_test_scaled=scaler.transform(x_test)
```

```python
x_test_scaled=pd.DataFrame(x_test_scaled,columns=x_test.columns)
x_train_scaled=pd.DataFrame(x_train_scaled,columns=x_train.columns)
```

```python
x_test_scaled.head(5)
```

```python
x_train_scaled.head(3)
```

## Graph of Before Scaling and After scaling the data

```python
fig,(ax1,ax2)=plt.subplots(1,2,figsize=(12,5))

ax1.set_title('Before Scalling')
ax1.scatter(x_train['Age'],x_train['EstimatedSalary'],c=y_train)

ax2.set_title('Before Scalling')
ax2.scatter(x_train_scaled['Age'],x_train_scaled['EstimatedSalary'],c=y_train)

plt.show()
```
